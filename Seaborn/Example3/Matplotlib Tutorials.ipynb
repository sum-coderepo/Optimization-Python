Matplotlib subplots Example

```python
fig, ax = plt.subplots(1,2)
ax[0].plot(x,y, 'b')
ax[1].plot(x,y, 'g')
```

Subplots with forloop


```python
fig, ax = plt.subplots(1,2)
col = ['r', 'g']
for i,axes in enumerate(ax):
    axes.plot(x,y,col[i])
```

Increase the figure size

```python
fig , ax = plt.subplots(figsize = (10,6), dpi = 100 )
ax.plot(x,y, 'r', label = 'y*x')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.legend()
ax.set_title('Random Number plot')
```

If we want to create a single figure with multiple lines, we can simply call the plot
function multiple times

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x));
```

Adjusting the plot colors and styles

```python
plt.plot(x, np.sin(x - 0), color='blue') # specify color by name
plt.plot(x, np.sin(x - 1), color='g') # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75') # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44') # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 and 1
plt.plot(x, np.sin(x - 5), color='chartreuse'); # all HTML color names supported
```

Adjust Line Plotting

```python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');
# For short, you can use the following codes:
plt.plot(x, x + 4, linestyle='-') # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':'); # dotted
```

Adjusting the plot Axes limit

```python
plt.plot(x, np.sin(x))
plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5);
```

The plt.axis() method allows you to set the x
and y limits with a single call, by passing a list that specifies [xmin, xmax, ymin,ymax]

To Plot one Unit of X is one unit of Y

```python
plt.plot(x, np.sin(x))
plt.axis('equal');
```

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')
plt.legend();
```

Instead of calling each function individually, we can set all properties at once

```python
ax = plt.axes()
ax.plot(x, np.sin(x))
ax.set(xlim=(0, 10), ylim=(-2, 2),
xlabel='x', ylabel='sin(x)',
title='A Simple Plot');
```

Basic Scatter Plot

```python
rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5), rng.rand(5), marker,
        label="marker='{0}'".format(marker))
```

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 100 * rng.rand(100)
plt.scatter(x, y , c = colors, s= sizes, alpha = 0.3, cmap = 'viridis')
plt.colorbar() # Show Color ScaleA
```

Basic Error bar

```python
x = np.linspace(0, 10, 50)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(50)
plt.errorbar(x, y, yerr=dy, fmt='.k');
```
